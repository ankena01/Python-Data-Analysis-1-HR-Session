```python
## Draw Line Graph using matplotlib and decorate it with using various elements

import matplotlib.pyplot as plt
import numpy as np

# Generating sample data
x = np.arange(1, 11)  # X-axis points
y = np.random.rand(10) * 100  # Y-axis points generated randomly

# Creating the line graph
plt.figure(figsize=(10, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Sample Data')

# Adding various elements
plt.title('Sample Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
## Draw bar graph using matplotlib and decorate it by adding various elements


# Generating sample data for the bar graph
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 33]

# Creating the bar graph
plt.figure(figsize=(8, 6))
bars = plt.bar(categories, values, color='cyan', edgecolor='black')

# Decorating the bar graph
plt.title('Sample Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(rotation=45)  # Rotating category labels for better visibility
plt.yticks(np.arange(0, 81, 10))

# Adding value labels above each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), va='bottom')  # va: vertical alignment

# Show the plot
plt.show()
```

```python
## Draw pie chart using matplotlib and decorate it by adding various elements

import matplotlib.pyplot as plt

# Data to plot
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [215, 130, 245, 210]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0.1, 0, 0, 0)  # explode 1st slice (Apple)

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Adding decoration
plt.title('Fruit Consumption Pie Chart')
plt.legend(labels, title="Fruits", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the plot
plt.show()
```

```python
# Explain CRUD operations in MongoDB with Example

from pymongo import MongoClient

# Connecting to the MongoDB server
client = MongoClient('mongodb://localhost:27017/')
db = client['school']  # Assuming the database name is 'school'

# Create operation: Adding a new student
student = {
    "name": "John Doe",
    "age": 20,
    "subjects": ["Mathematics", "Science"]
}
students_collection = db['students']
insert_result = students_collection.insert_one(student)
print(f"Inserted Student ID: {insert_result.inserted_id}")

# Read operation: Fetching a student by name
student_data = students_collection.find_one({"name": "John Doe"})
print(f"Student Data: {student_data}")

# Update operation: Updating student's age
update_result = students_collection.update_one({"name": "John Doe"}, {"$set": {"age": 21}})
print(f"Documents Updated: {update_result.modified_count}")

# Delete operation: Removing a student
delete_result = students_collection.delete_one({"name": "John Doe"})
print(f"Documents Deleted: {delete_result.deleted_count}")
```

```python
## Create a series from a NumPy array and find the max and mean of unique items of the series.
import pandas as pd
# Creating a NumPy array
np_array = np.array([1, 2, 2, 3, 4, 4, 5, 5, 5])

# Creating a Series from the NumPy array
series = pd.Series(np_array)
print("Series:")
print(series)

# Finding max and mean of unique items
unique_items = series.unique()
max_value = unique_items.max()
mean_value = unique_items.mean()
print(f"Unique items in series : {unique_items}")
print(f"Max of Unique Items: {max_value}")
print(f"Mean of Unique Items: {mean_value}")
```
